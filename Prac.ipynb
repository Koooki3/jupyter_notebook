```python
import numpy as np
import matplotlib.pyplot as plt
import torch
```

```python
# tensor create
x = torch.tensor([1, 2, 3])
print(x)
```

```python
# tensor opertion
x = torch.zeros(4, 3, dtype = torch.float32)
x = torch.randn_like(x, dtype = torch.float32)
print(x)

print(x.size())
print(x.shape)
```

```python
# random tensor setup
a = torch.IntTensor(1,2)
b = torch.FloatTensor(2,3)
print(a, "\n", b)
```

```python
# tensor numpy translation
g = np.array([[1,2,3],[4,5,6]])
h = torch.tensor(g)
print(h)
i = torch.from_numpy(g)
print(i)
j = h.numpy()
print(j)
```

```python
# tensor view
x = torch.randn(4, 4)
y = x.view(16)
z = x.view(-1, 8)
z2 = x.view(8, -1)
print(x.size(), "\n", y.size(), "\n", z.size(), "\n", z2.size())
# tensor operation
a = torch.ones(2, 2)
b = torch.ones(2, 2)
print(a + b)
print(a * b)
print(a / b)
print(a % b)
print(a ** b)
```
