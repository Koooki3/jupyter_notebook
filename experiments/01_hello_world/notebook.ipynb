## 简单计算示例

```python
x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)
df = pd.DataFrame({"x": x, "sin(x)": y})
df.head(10)
```

## 简单绘图示例

```python
plt.figure(figsize=(8, 4))
plt.plot(x, y, label="sin(x)")
plt.xlabel("x")
plt.ylabel("y")
plt.title("正弦曲线 - 环境测试")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```
